### remove nan values on Age 


```python
for i in range(df.shape[0]):
    if np.isnan(df.loc[i, 'Age']):
        if df.loc[i, 'Sex'] == 'female':
            df.loc[i, 'Age'] = 27.
        elif df.loc[i, 'Sex'] == 'male':
            df.loc[i, 'Age'] = 30.
    df.loc[i, 'Cabin'] = ord(df.loc[i, 'Cabin'])
df.head()
```

```python
y_values = df['Survived']
df = df.drop(['Survived'], axis=1)
df = df.drop(['PassengerId'], axis=1)
df = df.drop(['Name'], axis=1)
df = df.drop(['Ticket'], axis=1)
y_values.head()
```

## define neural network


```python
from torch.utils.data import Dataset, DataLoader

class TitanicDataset(Dataset):
    def __init__(self, array_element, array_label, train = True):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        if train == True:
            partiel = int(len(array_element)*0.7)
            
            self.dataset_data  = array_element[0: partiel]
            self.dataset_label = array_label[0: partiel]
            
        else:
            partiel = int(len(array_element)*0.3)
            self.dataset_data  = array_element[partiel : -1]
            self.dataset_label = array_label[partiel : -1]
        
    def __len__(self):
        return len(self.dataset_data)

    def __getitem__(self, idx):
        
        if torch.is_tensor(idx):
            idx = idx.tolist()

        data = self.dataset_data[idx, :]
        
        label = self.dataset_label[idx]
        sample = [torch.tensor(data), torch.tensor(label) ]
        
        return sample
```

```python
trainset = TitanicDataset(df.values, y_values.values, True)
trainloader = torch.utils.data.DataLoader(trainset, batch_size = 64, shuffle=True)

testset = TitanicDataset(df.values, y_values.values, False)
testloader = torch.utils.data.DataLoader(testset, batch_size = 64, shuffle=True)

train_sample_ = trainset[0]
print(train_sample_)

test_sample_ = testset[0]
print(test_sample_)
```

```python
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

#define model 
class model_titanic(nn.Module):
    def __init__(self):
        ##layer definition
        super(model_titanic, self).__init__()
        
        #MLP 
        self.fc1 = nn.Linear(8, 8)
        self.fc2 = nn.Linear(8, 8)
        self.fc3 = nn.Linear(8, 2)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))

        x = F.relu(self.fc2(x))

        x = F.relu(self.fc3(x))
        
        return x

##
model = model_titanic()
    
#define error mesurment
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

#train network
n_epoch = 1000
valid_loss_min = []
valid_loss_min_test = []

##train
for epoch in range(n_epoch):
    train_loss = 0.0
    model.train()
    for data, target in trainloader:
        optimizer.zero_grad()
        
        output = model(data.float())
        
        loss = criterion(output, target)
        
        loss.backward()
        
        optimizer.step()
        
        train_loss += loss.item()/data.size()[0]
    valid_loss_min.append( 1- train_loss)
    
    test_loss = 0.0
    
    for data, target in testloader:
        model.eval()

        output = model(data.float())

        loss = criterion(output, target)
        
        test_loss += loss.item()/data.size()[0]
        
    valid_loss_min_test.append( 1 - test_loss )
plt.figure()
plt.plot(np.array(valid_loss_min))
plt.plot( np.array( valid_loss_min_test ))
```
